## Abstract

El presente análisis busca explorar los factores que influyen en el rendimiento académico de los estudiantes a partir de un dataset con 10,000 observaciones. Se analizan variables como horas de estudio, uso de tecnología educativa, nivel de estrés, asistencia y estilo de aprendizaje. 

El objetivo principal es identificar patrones que puedan ayudar a mejorar la educación y determinar qué factores tienen mayor impacto en los puntajes de examen. Se plantean las siguientes preguntas: 
1. ¿Más horas de estudio garantizan mejores calificaciones?  
2. ¿Existe una relación entre el nivel de estrés y los puntajes de examen?  
3. ¿El uso de tecnología educativa tiene un impacto positivo en los resultados?  

A través del análisis exploratorio de datos (EDA) se realizan visualizaciones y estadísticas descriptivas para encontrar relaciones significativas entre estas variables.

## Preguntas de interés e Hipótesis

### 1 ¿Más horas de estudio garantizan mejores calificaciones?
**Hipótesis:** Los estudiantes que estudian más horas a la semana obtienen mejores puntajes en los exámenes.

### 2️ ¿Existe una relación entre el nivel de estrés y los puntajes de examen?
**Hipótesis:** Los estudiantes con un nivel de estrés más alto tienden a obtener menores puntajes en los exámenes.

### 3️ ¿El uso de tecnología educativa tiene un impacto positivo en los resultados?
**Hipótesis:** Los estudiantes que usan herramientas educativas en línea obtienen mejores calificaciones en promedio.


```python
import pandas as pd

df = pd.read_csv("student_performance_large_dataset.csv")

# Ver primeras filas
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df["Exam_Score (%)"], bins=20, kde=True)
plt.title("Distribución de los Puntajes de Examen")
plt.xlabel("Puntaje (%)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x="Self_Reported_Stress_Level", y="Exam_Score (%)", data=df)
plt.title("Distribución de Puntajes según Nivel de Estrés")
plt.xlabel("Nivel de Estrés Reportado")
plt.ylabel("Puntaje (%)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x="Gender", y="Exam_Score (%)", data=df)
plt.title("Comparación de Puntajes de Examen por Género")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
plt.hexbin(df["Study_Hours_per_Week"], df["Exam_Score (%)"], gridsize=30, cmap="Blues", alpha=0.3)
plt.colorbar(label="Cantidad de Estudiantes")
plt.title("Relación entre Horas de Estudio y Puntaje (Mapa de Densidad)")
plt.xlabel("Horas de Estudio por Semana")
plt.ylabel("Puntaje (%)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.violinplot(x="Use_of_Educational_Tech", y="Exam_Score (%)", data=df)
plt.title("Distribución de Puntajes según Uso de Tecnología Educativa")
plt.xlabel("Uso de Tecnología Educativa")
plt.ylabel("Puntaje (%)")
plt.show()



```

## Conclusión Final del Análisis 📊

Después de explorar los datos y realizar varias visualizaciones, llegue a las siguientes conclusiones sobre los factores que influyen en el rendimiento académico de los estudiantes.

## Distribución de Puntajes de Examen
📌 Observación:

La mayoría de los puntajes están distribuidos de manera uniforme entre 40% y 100%, sin un sesgo claro hacia puntajes muy altos o muy bajos.
Los estudiantes que obtienen 100% parecen ser un poco más frecuentes que otros puntajes.
Este gráfico nos dice que no hay una "mayoría" de estudiantes con notas excelentes o muy bajas, sino que están repartidos en distintos niveles de rendimiento.
✅ Conclusión:
Los puntajes de los estudiantes son diversos y no hay una tendencia fuerte hacia puntajes extremadamente altos o bajos.

## Estrés y Rendimiento Académico
📌 Observación:

Al comparar los puntajes de los estudiantes según su nivel de estrés (High, Medium, Low), se observa que las distribuciones son casi idénticas.
La mediana y los valores extremos de cada grupo no muestran diferencias significativas.
✅ Conclusión:
El nivel de estrés no parece afectar de manera significativa los puntajes de examen.
Aunque el estrés es un factor relevante en la educación, en este caso no se observó una relación clara entre el estrés y el rendimiento académico.

## Horas de Estudio vs Puntajes de Examen
📌 Observación:

Los estudiantes estudian entre 10 y 40 horas por semana en su mayoría.
No hay una tendencia clara que indique que más horas de estudio se traducen en mejores puntajes.
Algunos estudiantes con pocas horas de estudio obtienen notas altas, mientras que otros con muchas horas de estudio no necesariamente tienen mejores resultados.
✅ Conclusión:
El número de horas de estudio no es un factor determinante en los puntajes de examen.
Esto sugiere que la calidad del estudio puede ser más importante que la cantidad de horas dedicadas.

## Uso de Tecnología Educativa y Rendimiento Académico
📌 Observación:

Se compararon los puntajes de estudiantes que usan y no usan tecnología educativa.
Los promedios de ambos grupos son casi iguales:
No usan tecnología: 70.33%
Sí usan tecnología: 70.12%
La variabilidad en los puntajes (desviación estándar) también es similar en ambos casos.
✅ Conclusión:
El uso de tecnología educativa no tiene un impacto significativo en los puntajes de examen.
Este hallazgo sugiere que el simple acceso a la tecnología no mejora automáticamente el rendimiento académico. Otros factores, como métodos de enseñanza, estrategias de estudio y motivación, podrían ser más influyentes.
